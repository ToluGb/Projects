```python
import cv2
import sys
import dlib
import argparse
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
import io 

img = np.genfromtxt(io.BytesIO(uploaded['input00.txt']), delimiter=",", skip_header=1)
print(img.shape)
img = img.astype(np.uint8)
```

```python
# Show the image 
plt.figure (figsize=(20,20))
plt.imshow(img, cmap=plt.get_cmap('gray'))
plt.show()
```

```python
# Get user supplied value
imagePath = sys.argv[1]
```

```python
# Now we create the cascade and initialize it with our face cascade
faceCascade = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_default.xml")
```

```python
# Read the image
cv2.imread(imagePath)
```

```python
# Convert the image to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BAYER_BG2GRAY)
```

```python
# Detect faces in the image
faces = faceCascade.detectMultiScale(
    gray,
    scaleFactor=1.1,
    minNeighbors=5,
    minSize=(30, 30),
    flags = cv2.CASCADE_SCALE_IMAGE
)
```

```python
print ("Found {} faces!".format(len(faces)))

# Draw a rectangle around the faces
for (i, rect) in enumerate(faces):
    x = rect.left()
    y = rect.top()
    w = rect.right() - x
    h = rect.bottom() - y
    #draw a rectangle
    cv2.rectangle(img, (x, y), (x + w, y + h), (0, 255, 0), 2)
```
